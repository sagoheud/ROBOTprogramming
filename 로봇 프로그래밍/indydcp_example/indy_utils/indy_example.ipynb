```python
# Indy2 2adm

from indy_utils import indydcp_client as client
import time
robot_ip = "192.168.3.2"
robot_name = "NRMK-Indy7"
indy = client.IndyDCPClient(robot_ip, robot_name)

indy.connect()
PICK_BASE = [0.197853293248411, 0.494934342906699, 0.1533289339019412] # 픽 기준점
PLACE_BASE = [0.14605595778978153, 0.30356181695104195, 0.19940150110518265] # 플레이스 기준점
GRID_X = 1                      # 가로 1칸
GRID_Y = 2                      # 세로 2칸
NUM_LAYERS = 2                  # 총 수
LAYER_HEIGHT = 0.03             # 층 높이 (m)
OFFSET_X = 0.04                 # X 간격
OFFSET_Y = 0.04                 # Y 간격
RETRACT_Z = 0.1                # Pick/Place 리트랙
ROTATION = [-22.51139089331165, -174.81963954473787, 66.80689867536479]          # 자세 (rx, ry, rz)
ROTATION1 = [-0.006518109219459515, -179.98968930569117, 120.2199313262332]

# 좌표 값 체크 함수
def move_done_check():
    while True:
        status = indy.get_robot_status()
        if status['movedone'] == 1:
            break
        time.sleep(0.2)

# 팔레타이징 좌표 생성 함수
def generate_grid(base, grid_x, grid_y, offset_x, offset_y, num_layers, layer_height):
    coords = []
    for layer in range(num_layers):
        z = base[2] + layer * layer_height
        for i in range(grid_y):
            for j in range(grid_x):
                x = base[0] + j * offset_x
                y = base[1] + i * offset_y
                coords.append([x,y,z])
                print(f"생성된 좌표 (층 {layer+1}): {coords[-1]}")
    return coords

# 자재 들어오는 센서 신호 대기 함수
def wait_for_sensor(di_pin):
    print(f"[{di_pin}번 센서] 신호 대기 중...")
    while True:
        # get_di는 리스트 형태나 특정 구조로 반환될 수 있으니 
        # 사용 중인 라이브러리 버전에 따라 인덱스를 확인하세요.
        di_status = indy.get_di()
        
        # 일반적으로 di_status[di_pin] 형식을 사용합니다.
        if di_status[di_pin] == 1: 
            print(f"[{di_pin}번 센서] 감지 완료!")
            break
        time.sleep(0.1) # CPU 부하를 줄이기 위한 짧은 대기

# 픽엔플레이스 작동 함수
def execute_pick_and_place():
    indy.go_home()
    move_done_check()
    print("\n홈 위치 도착 완료.")
    time.sleep(1)
    place_positions = generate_grid(PLACE_BASE, GRID_X, GRID_Y, OFFSET_X, OFFSET_Y, 
                                    NUM_LAYERS, LAYER_HEIGHT)
    print("\n픽 고정 위치:", PICK_BASE)
    print("팔레타이징 위치 수:", len(place_positions))

    # 첫 번째 픽 접근
    indy.task_move_to([PICK_BASE[0], PICK_BASE[1], PICK_BASE[2] + RETRACT_Z, *ROTATION])
    move_done_check()
    for i, place in enumerate(place_positions):
        wait_for_sensor(3)                  # 센서 신호 대기
        print(f"\n===== Step {i + 1} =====")
        print(f"Pick : {PICK_BASE}")
        print(f"Place: {place}")
        indy.task_move_to([PICK_BASE[0], PICK_BASE[1], PICK_BASE[2] + RETRACT_Z, *ROTATION])
        move_done_check()
        indy.task_move_to([PICK_BASE[0], PICK_BASE[1], PICK_BASE[2], *ROTATION])
        move_done_check()
        indy.set_do(2, True)
        print("진공 ON (흡착 중...)")
        time.sleep(1)
        indy.task_move_to([PICK_BASE[0], PICK_BASE[1], PICK_BASE[2] + RETRACT_Z, *ROTATION])
        move_done_check()
        indy.task_move_to([place[0], place[1], place[2] + RETRACT_Z, *ROTATION1])
        move_done_check()
        indy.task_move_to([place[0], place[1], place[2], *ROTATION1])
        move_done_check()
        indy.set_do(2, False)
        print("진공 OFF (해제 중...)")
        time.sleep(1)
        indy.task_move_to([place[0], place[1], place[2] + RETRACT_Z, *ROTATION1])
        move_done_check()
    indy.go_home()
    move_done_check()
    print("\n모든 Pick & Place 작업 완료!")

execute_pick_and_place()
indy.disconnect()
print("로봇 연결 종료.")
```

```python
from indy_utils import indydcp_client as client
import time
import serial

arduino = serial.Serial('COM13', 9600, timeout=1)
time.sleep(2)
robot_ip = "192.168.3.2"
robot_name = "NRMK-Indy7"
indy = client.IndyDCPClient(robot_ip, robot_name)
indy.connect()
print("로봇 연결 성공")

def lcd_send(msg):
    arduino.write((msg + "\n").encode())

def move_done_check():
    while True:
        if indy.get_robot_status()['movedone'] == 1:
            break
        time.sleep(0.1)

def generate_grid(base, grid_x, grid_y, offset_x, offset_y, layers=1, layer_h=0):
    coords = []
    for layer in range(layers):
        z = base[2] + layer * layer_h
        for i in range(grid_y):
            for j in range(grid_x):
                x = base[0] + j * offset_x
                y = base[1] + i * offset_y
                coords.append([x, y, z])
    return coords

# 작업1 (2by2 Pick & Place)
def task_pick2by2_place2by2():
    lcd_send("OP1_START")
    print("\n=== 작업1 시작 ===")
    PICK_BASE = [0.18605595778978153, 0.34356181695104195, 0.19940150110518265] # 픽 기준점
    PLACE_BASE = [0.06905595778978153, 0.22356181695104195, 0.19940150110518265] # 플레이스 기준점
    PICK_GRID_X = 2
    PICK_GRID_Y = 2
    PLACE_GRID_X = 2
    PLACE_GRID_Y = 2
    OFFSET_X = 0.04
    OFFSET_Y = 0.04
    RETRACT_Z = 0.05
    ROT_P = [-0.006518109219459515, -179.98968930569117, 120.2199313262332]
    ROT_R = [-0.006518109219459515, -179.98968930569117, 120.2199313262332]

    pick_positions = generate_grid(PICK_BASE, PICK_GRID_X, PICK_GRID_Y, OFFSET_X, OFFSET_Y)
    place_positions = generate_grid(PLACE_BASE, PLACE_GRID_X, PLACE_GRID_Y, OFFSET_X, OFFSET_Y)

    indy.go_home(); move_done_check()

    for i in range(len(pick_positions)):
        pick = pick_positions[i]
        place = place_positions[i]
        print(f">> STEP {i+1}")

        indy.task_move_to([pick[0], pick[1], pick[2] + RETRACT_Z, *ROT_P]); move_done_check()
        indy.task_move_to([pick[0], pick[1], pick[2], *ROT_P]); move_done_check()
        indy.set_do(2, True); time.sleep(0.5)
        indy.task_move_to([pick[0], pick[1], pick[2] + RETRACT_Z, *ROT_P]); move_done_check()

        indy.task_move_to([place[0], place[1], place[2] + RETRACT_Z, *ROT_R]); move_done_check()
        indy.task_move_to([place[0], place[1], place[2], *ROT_R]); move_done_check()
        indy.set_do(2, False); time.sleep(0.5)
        indy.task_move_to([place[0], place[1], place[2] + RETRACT_Z, *ROT_R]); move_done_check()

    indy.go_home(); move_done_check()

    lcd_send("OP1_FINISH")
    print("작업1 종료")

# 작업2 (2by2 * 2단 팔레타이징)
def task_pal_1by2_2layer():
    lcd_send("OP2_START")
    print("\n=== 작업2 시작 ===")

    PICK_BASE = [0.197853293248411, 0.494934342906699, 0.1533289339019412]
    PLACE_BASE = [0.11105595778978153, 0.26356181695104195, 0.19940150110518265]
    GRID_X = 2
    GRID_Y = 2
    NUM_LAYERS = 2
    OFFSET_X = 0.04
    OFFSET_Y = 0.04
    LAYER_H = 0.03
    RETRACT_Z = 0.1
    ROT_P = [-22.51139089331165, -174.81963954473787, 66.80689867536479]
    ROT_R = [-0.006518109219459515, -179.98968930569117, 120.2199313262332]

    place_positions = generate_grid(PLACE_BASE, GRID_X, GRID_Y, OFFSET_X, OFFSET_Y, NUM_LAYERS, LAYER_H)

    indy.go_home(); move_done_check()

    for i, place in enumerate(place_positions):
        print(f">> STEP {i+1}")

        indy.task_move_to([PICK_BASE[0], PICK_BASE[1], PICK_BASE[2] + RETRACT_Z, *ROT_P]); move_done_check()
        indy.task_move_to([PICK_BASE[0], PICK_BASE[1], PICK_BASE[2], *ROT_P]); move_done_check()
        indy.set_do(2, True); time.sleep(0.5)
        indy.task_move_to([PICK_BASE[0], PICK_BASE[1], PICK_BASE[2] + RETRACT_Z, *ROT_P]); move_done_check()

        indy.task_move_to([place[0], place[1], place[2] + RETRACT_Z, *ROT_R]); move_done_check()
        indy.task_move_to([place[0], place[1], place[2], *ROT_R]); move_done_check()
        indy.set_do(2, False); time.sleep(0.5)
        indy.task_move_to([place[0], place[1], place[2] + RETRACT_Z, *ROT_R]); move_done_check()
    indy.go_home(); move_done_check()

    lcd_send("OP2_FINISH")
    print("작업2 종료")
    
print("\n=== 아두이노우노 인디7 뉴로메카로봇 ===")
while True:
    if arduino.in_waiting:
        cmd = arduino.readline().decode().strip()
        if cmd == "OP1":
            task_pick2by2_place2by2()
        elif cmd == "OP2":
            task_pal_1by2_2layer()
        elif cmd == "QUIT":
            lcd_send("PROCESS_QUIT")
            print("프로그램을 종료합니다.")
            indy.disconnect()
        break
    time.sleep(0.05)


    
```

```python
import customtkinter as ctk
import tkinter as tk
from time import sleep
import threading, queue, sys, math
import numpy as np
import matplotlib
matplotlib.use("TkAgg")
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

ctk.set_appearance_mode("dark")
ctk.set_default_color_theme("blue")

# ─────────────────────────────────────────
# Indy 연결
# ─────────────────────────────────────────
try:
    import indy_utils.indydcp_client as client
    ROBOT_IP, ROBOT_NAME = "192.168.3.2", "NRMK-Indy7"
    indy = client.IndyDCPClient(ROBOT_IP, ROBOT_NAME)
    SIMULATION_MODE = False
except ImportError:
    SIMULATION_MODE = True
    indy = None

# ─────────────────────────────────────────
# DH 파라미터  [a, d, alpha, theta_offset]  단위: m
# 홈 실측: [~0, ~0, -90, ~0, -90, ~0] deg
# ─────────────────────────────────────────
INDY7_DH = [
    [0,      0.3650,  math.pi/2,  0],
    [0.3220, 0,       0,          0],
    [0,      0,       math.pi/2,  0],
    [0,      0.3320, -math.pi/2,  0],
    [0,      0,       math.pi/2,  0],
    [0,      0.1500,  0,          0],
]
HOME_ANGLES_DEG = [0.0, 0.0, -90.0, 0.0, -90.0, 0.0]
ZERO_ANGLES_DEG = [0.0]*6

# 링크별 표현 굵기 (linewidth) — 베이스→EE 순
LINK_LW    = [10, 8, 7, 6, 5, 4]
LINK_COLS  = ['#1E7FBF','#1E9FDF','#1AAFDF','#16BFDF','#12CFDF','#0DDFEF']
JOINT_COLS = ['#AABBCC','#F07030','#F07030','#F07030','#F07030','#F07030','#F07030']
JOINT_SZ   = [120, 80, 70, 60, 50, 40, 30]

# ─────────────────────────────────────────
# 순기구학
# ─────────────────────────────────────────
def dh_transform(a, d, alpha, theta):
    ct, st = math.cos(theta), math.sin(theta)
    ca, sa = math.cos(alpha), math.sin(alpha)
    return np.array([
        [ct, -st*ca,  st*sa, a*ct],
        [st,  ct*ca, -ct*sa, a*st],
        [0,   sa,     ca,    d   ],
        [0,   0,      0,     1   ]
    ])

def forward_kinematics(joint_angles_deg):
    T = np.eye(4)
    pos = [np.zeros(3)]
    for i, (a, d, alpha, offset) in enumerate(INDY7_DH):
        theta = math.radians(joint_angles_deg[i]) + offset
        T = T @ dh_transform(a, d, alpha, theta)
        pos.append(T[:3, 3].copy())
    return pos, T

# ─────────────────────────────────────────
# 링크 튜브 윤곽선 생성 (가벼운 방식)
# 원기둥을 측면 N개 선으로 근사 → plot()만 사용
# ─────────────────────────────────────────
def tube_lines(p0, p1, radius, n=6):
    """원기둥 윤곽을 n쌍의 선분 좌표로 반환 (xs, ys, zs 리스트)"""
    axis = p1 - p0
    length = np.linalg.norm(axis)
    if length < 1e-6:
        return []
    zhat = axis / length
    tmp  = np.array([1,0,0]) if abs(zhat[0]) < 0.9 else np.array([0,1,0])
    xhat = np.cross(zhat, tmp);  xhat /= np.linalg.norm(xhat)
    yhat = np.cross(zhat, xhat)

    lines = []
    for t in np.linspace(0, 2*math.pi, n, endpoint=False):
        offset = (math.cos(t)*xhat + math.sin(t)*yhat) * radius
        b = p0 + offset
        tp = p1 + offset
        lines.append((
            [b[0], tp[0]],
            [b[1], tp[1]],
            [b[2], tp[2]]
        ))
    return lines

def ring_line(center, radius, normal, n=16):
    """관절 링(원) 하나를 폴리라인으로"""
    tmp  = np.array([1,0,0]) if abs(normal[0]) < 0.9 else np.array([0,1,0])
    xhat = np.cross(normal, tmp);  xhat /= np.linalg.norm(xhat)
    yhat = np.cross(normal, xhat)
    t    = np.linspace(0, 2*math.pi, n+1)
    pts  = np.array([center + radius*(math.cos(ti)*xhat + math.sin(ti)*yhat) for ti in t])
    return pts[:,0], pts[:,1], pts[:,2]

# ─────────────────────────────────────────
# 전역 상태
# ─────────────────────────────────────────
stop_flag         = False
manual_mode       = False
joint_sliders     = []
angle_labels      = []
gauge_bars        = []
canvas_3d         = None
ax_3d             = None
fig_3d            = None
status_label      = None
ee_info_label     = None
mode_indicator    = None
manual_btn        = None
robot_angle_queue = queue.Queue(maxsize=1)

# ─────────────────────────────────────────
# 로그 핸들러
# ─────────────────────────────────────────
class TextHandler:
    def __init__(self, w):
        self.w = w
    def write(self, s):
        if s.strip():
            self.w.configure(state="normal")
            self.w.insert(tk.END, s if s.endswith("\n") else s+"\n")
            self.w.see(tk.END)
            self.w.configure(state="disabled")
    def flush(self): pass

# ─────────────────────────────────────────
# 3D 렌더링  ★ 메인스레드 전용, 가볍게
# ─────────────────────────────────────────
def draw_robot_3d(joint_angles_deg):
    if ax_3d is None or canvas_3d is None:
        return

    ax_3d.cla()
    ax_3d.set_facecolor('#06090F')
    fig_3d.patch.set_facecolor('#06090F')

    ax_3d.set_xlim(-1.0, 1.0)
    ax_3d.set_ylim(-1.0, 1.0)
    ax_3d.set_zlim( 0.0, 1.4)
    ax_3d.set_xlabel('X [m]', color='#3A5570', fontsize=7, labelpad=1)
    ax_3d.set_ylabel('Y [m]', color='#3A5570', fontsize=7, labelpad=1)
    ax_3d.set_zlabel('Z [m]', color='#3A5570', fontsize=7, labelpad=1)
    ax_3d.tick_params(colors='#223344', labelsize=6)
    ax_3d.xaxis.pane.fill = False
    ax_3d.yaxis.pane.fill = False
    ax_3d.zaxis.pane.fill = False
    ax_3d.xaxis.pane.set_edgecolor('#0D1520')
    ax_3d.yaxis.pane.set_edgecolor('#0D1520')
    ax_3d.zaxis.pane.set_edgecolor('#0D1520')
    ax_3d.grid(True, color='#0D1520', linewidth=0.4)

    pos, T_end = forward_kinematics(joint_angles_deg)

    # ── 베이스 플레이트 링
    br, n_b = 0.10, 20
    th  = np.linspace(0, 2*math.pi, n_b+1)
    ax_3d.plot(br*np.cos(th), br*np.sin(th), np.zeros(n_b+1),
               color='#8899AA', linewidth=3, alpha=0.85)
    # 베이스 십자선
    for dx, dy in [(br,0),(-br,0),(0,br),(0,-br)]:
        ax_3d.plot([0,dx],[0,dy],[0,0], color='#556677', linewidth=1.5, alpha=0.6)

    # ── 링크: 중심선(굵) + 튜브 윤곽선(얇)
    for i in range(len(pos)-1):
        p0, p1 = pos[i], pos[i+1]
        col = LINK_COLS[i]
        lw  = LINK_LW[i]

        # 중심선 (굵게, 메인 형태)
        ax_3d.plot([p0[0],p1[0]], [p0[1],p1[1]], [p0[2],p1[2]],
                   color=col, linewidth=lw, alpha=0.9, solid_capstyle='round')
        # 글로우
        ax_3d.plot([p0[0],p1[0]], [p0[1],p1[1]], [p0[2],p1[2]],
                   color=col, linewidth=lw*2.2, alpha=0.10)

        # 튜브 윤곽 (가는 선 4개로 원기둥 실루엣)
        r = 0.010 + 0.006*(5-i)   # 베이스→EE 방향으로 감소
        for xs, ys, zs in tube_lines(p0, p1, r, n=4):
            ax_3d.plot(xs, ys, zs, color=col, linewidth=0.8, alpha=0.35)

    # ── 관절: scatter + 링
    for i, p in enumerate(pos):
        sz  = JOINT_SZ[i]
        col = JOINT_COLS[i]
        ax_3d.scatter([p[0]], [p[1]], [p[2]], s=sz, c=col,
                      depthshade=True, zorder=5, edgecolors='#FFFFFF22', linewidths=0.5)
        # 관절 링 (링크 축 방향)
        if i < len(pos)-1:
            axis = pos[i+1] - p
            ln   = np.linalg.norm(axis)
            if ln > 1e-6:
                rx, ry, rz = ring_line(p, 0.012 + 0.005*(5-i), axis/ln, n=12)
                ax_3d.plot(rx, ry, rz, color='#FFFFFF', linewidth=0.8, alpha=0.4)

    # ── EE 좌표축
    ee = pos[-1]
    R  = T_end[:3,:3]
    sc = 0.08
    for ax_i, (col, lbl) in enumerate(zip(['#FF3333','#33FF44','#3366FF'],['X','Y','Z'])):
        d = R[:, ax_i] * sc
        ax_3d.quiver(ee[0], ee[1], ee[2], d[0], d[1], d[2],
                     color=col, linewidth=2, arrow_length_ratio=0.3)

    # ── EE 마커 + 텍스트
    ax_3d.scatter([ee[0]], [ee[1]], [ee[2]], c='#FF2255', s=50, zorder=9,
                  edgecolors='#FFFFFF55', linewidths=1)
    ax_3d.text(ee[0], ee[1], ee[2]+0.09,
               f"EE ({ee[0]:.3f}, {ee[1]:.3f}, {ee[2]:.3f})",
               color='#CCDDF0', fontsize=6, ha='center',
               bbox=dict(boxstyle='round,pad=0.25', facecolor='#0A1525',
                         alpha=0.85, edgecolor='#0099BB', linewidth=0.8))

    # ── 관절 번호
    jnames = ['Base','J1','J2','J3','J4','J5','J6']
    for i, p in enumerate(pos):
        off = 0.04 + 0.01*i
        ax_3d.text(p[0], p[1], p[2]+off, jnames[i],
                   color='#FF9944', fontsize=5.5, ha='center', fontweight='bold')

    ax_3d.set_title('Indy7  실시간 3D', color='#00BBDD',
                    fontsize=9, pad=5, fontweight='bold')

    if ee_info_label:
        ee_info_label.configure(
            text=f"EE  X={ee[0]:+.4f} m    Y={ee[1]:+.4f} m    Z={ee[2]:+.4f} m")

    canvas_3d.draw_idle()

# ─────────────────────────────────────────
# 메인스레드 큐 루프
# ─────────────────────────────────────────
def process_robot_queue():
    if not manual_mode:
        try:
            angles_deg = robot_angle_queue.get_nowait()
            for i, s in enumerate(joint_sliders):
                if i < len(angles_deg):
                    s.set(angles_deg[i])
                    angle_labels[i].configure(text=f"{angles_deg[i]:+7.2f}°")
                    gauge_bars[i].set((angles_deg[i]+180)/360)
            draw_robot_3d(angles_deg)
        except queue.Empty:
            pass
    root.after(120, process_robot_queue)

# ─────────────────────────────────────────
# 백그라운드 폴링 (큐에만 넣음)
# ─────────────────────────────────────────
def polling_thread():
    while True:
        if not SIMULATION_MODE and indy and not manual_mode:
            try:
                angles_deg = indy.get_joint_pos()   # degree 단위
                if angles_deg and len(angles_deg) == 6:
                    try:    robot_angle_queue.get_nowait()
                    except: pass
                    robot_angle_queue.put_nowait(list(angles_deg))
            except Exception:
                pass
        sleep(0.12)

# ─────────────────────────────────────────
# 유틸
# ─────────────────────────────────────────
def set_status(text, color="#FFFFFF"):
    try: status_label.configure(text=f"●  {text}", text_color=color)
    except: pass

def move_done_check():
    while True:
        if stop_flag: return False
        if SIMULATION_MODE: sleep(0.3); return True
        try:
            if indy.get_robot_status().get('movedone',0) == 1: return True
        except: pass
        sleep(0.08)

def generate_grid(base, gx, gy, ox, oy, layers=1, lh=0):
    out = []
    for layer in range(layers):
        z = base[2]+layer*lh
        for i in range(gy):
            for j in range(gx):
                out.append([base[0]+j*ox, base[1]+i*oy, z])
    return out

def run_task(fn):
    global stop_flag
    stop_flag = False
    threading.Thread(target=fn, daemon=True).start()

def _apply_angles_to_ui(angles_deg):
    def _do():
        for i, s in enumerate(joint_sliders):
            s.set(angles_deg[i])
            angle_labels[i].configure(text=f"{angles_deg[i]:+7.2f}°")
            gauge_bars[i].set((angles_deg[i]+180)/360)
        draw_robot_3d(angles_deg)
    root.after(0, _do)

# ─────────────────────────────────────────
# 슬라이더 콜백
# ─────────────────────────────────────────
def _on_slider(val, idx):
    angle_labels[idx].configure(text=f"{val:+7.2f}°")
    gauge_bars[idx].set((val+180)/360)
    draw_robot_3d([s.get() for s in joint_sliders])
    if manual_mode and not SIMULATION_MODE and indy:
        try: indy.joint_move_to([math.radians(s.get()) for s in joint_sliders])
        except: pass

def toggle_manual_mode():
    global manual_mode
    manual_mode = not manual_mode
    if manual_mode:
        manual_btn.configure(text="🔒  수동 ON  (폴링 정지)",
                             fg_color="#C05010", text_color="#FFFFFF")
        if mode_indicator: mode_indicator.configure(text="● MANUAL", text_color="#FF8C42")
        set_status("수동 슬라이더 모드", "#FF8C42")
    else:
        manual_btn.configure(text="🎚  수동 OFF  (실시간 폴링)",
                             fg_color="#1A2A50", text_color="#88BBFF")
        if mode_indicator: mode_indicator.configure(text="● POLLING", text_color="#44FF88")
        set_status("실시간 폴링 모드", "#44FF88")

# ─────────────────────────────────────────
# 로봇 동작
# ─────────────────────────────────────────
def stop_immediate():
    global stop_flag
    stop_flag = True
    if not SIMULATION_MODE and indy: indy.stop_motion()
    set_status("즉시 정지됨", "#FF3366")
    print("!!! 동작 즉시 정지 !!!")

def move_home():
    print("홈 위치로 이동 중...")
    set_status("홈 이동 중...", "#FFB347")
    if not SIMULATION_MODE and indy: indy.go_home()
    if move_done_check():
        print("홈 도착 완료"); set_status("홈 도착","#44FF88")
        _apply_angles_to_ui(HOME_ANGLES_DEG)

def move_zero():
    print("영점 위치로 이동 중...")
    set_status("영점 이동 중...", "#FFB347")
    if not SIMULATION_MODE and indy: indy.go_zero()
    if move_done_check():
        print("영점 도착 완료"); set_status("영점 도착","#44FF88")
        _apply_angles_to_ui(ZERO_ANGLES_DEG)

def task_pick2by2_place2by2():
    print("\n=== [작업1] 2x2 Pick & Place 시작 ===")
    set_status("작업1 실행 중","#00D4FF")
    PB  = [0.18605595778978153, 0.34356181695104195, 0.19940150110518265]
    PLB = [0.06905595778978153, 0.22356181695104195, 0.19940150110518265]
    OX,OY,RZ = 0.04,0.04,0.05
    ROT = [-0.006518109219459515,-179.98968930569117,120.2199313262332]
    pp = generate_grid(PB,2,2,OX,OY); pl = generate_grid(PLB,2,2,OX,OY)
    if not SIMULATION_MODE and indy: indy.go_home()
    if not move_done_check(): return
    for i in range(4):
        if stop_flag: break
        p,d = pp[i],pl[i]; print(f">> STEP {i+1}")
        if not SIMULATION_MODE and indy:
            indy.task_move_to([p[0],p[1],p[2]+RZ,*ROT]); move_done_check()
            indy.task_move_to([p[0],p[1],p[2],   *ROT]); move_done_check()
            indy.set_do(2,True);  sleep(0.5)
            indy.task_move_to([p[0],p[1],p[2]+RZ,*ROT]); move_done_check()
            if stop_flag: break
            indy.task_move_to([d[0],d[1],d[2]+RZ,*ROT]); move_done_check()
            indy.task_move_to([d[0],d[1],d[2],   *ROT]); move_done_check()
            indy.set_do(2,False); sleep(0.5)
            indy.task_move_to([d[0],d[1],d[2]+RZ,*ROT]); move_done_check()
        else: sleep(0.4)
    if not SIMULATION_MODE and indy: indy.go_home(); move_done_check()
    print("작업1 종료"); set_status("대기 중","#44FF88")

def task_pal_2by2_2layer():
    print("\n=== [작업2] 팔레타이징 시작 ===")
    set_status("작업2 실행 중","#FFD700")
    PB  = [0.197853293248411,0.494934342906699,0.1533289339019412]
    PLB = [0.11105595778978153,0.26356181695104195,0.19940150110518265]
    OX,OY,LH,RZ = 0.04,0.04,0.03,0.1
    RP_ = [-22.51139089331165,-174.81963954473787,66.80689867536479]
    RR  = [-0.006518109219459515,-179.98968930569117,120.2199313262332]
    plc = generate_grid(PLB,2,2,OX,OY,2,LH)
    if not SIMULATION_MODE and indy: indy.go_home()
    if not move_done_check(): return
    for i,pl in enumerate(plc):
        if stop_flag: break
        print(f">> STEP {i+1}")
        if not SIMULATION_MODE and indy:
            indy.task_move_to([PB[0],PB[1],PB[2]+RZ,*RP_]); move_done_check()
            indy.task_move_to([PB[0],PB[1],PB[2],   *RP_]); move_done_check()
            indy.set_do(2,True); sleep(0.5)
            indy.task_move_to([PB[0],PB[1],PB[2]+RZ,*RP_]); move_done_check()
            if stop_flag: break
            indy.task_move_to([pl[0],pl[1],pl[2]+RZ,*RR]); move_done_check()
            indy.task_move_to([pl[0],pl[1],pl[2],   *RR]); move_done_check()
            indy.set_do(2,False); sleep(0.5)
            indy.task_move_to([pl[0],pl[1],pl[2]+RZ,*RR]); move_done_check()
        else: sleep(0.4)
    if not SIMULATION_MODE and indy: indy.go_home(); move_done_check()
    print("작업2 종료"); set_status("대기 중","#44FF88")

def shutdown():
    global stop_flag
    stop_flag = True
    if not SIMULATION_MODE and indy:
        try: indy.stop_motion()
        except: pass
        try: indy.disconnect()
        except: pass
    root.destroy()

# ═══════════════════════════════════════════════════════
#  GUI
# ═══════════════════════════════════════════════════════
root = ctk.CTk()
root.title("Indy7 로봇 제어 시스템")
root.geometry("1560x860")
root.configure(fg_color="#070C1A")

# ── 헤더
hdr = ctk.CTkFrame(root, fg_color="#0B1228", corner_radius=0, height=54)
hdr.pack(fill="x")
hdr.pack_propagate(False)
ctk.CTkLabel(hdr, text="  NEUROMEKA  ·  INDY7  CONTROL  SYSTEM",
             font=ctk.CTkFont("Courier New",17,"bold"),
             text_color="#00CCEE").pack(side="left", padx=20, pady=14)
status_label = ctk.CTkLabel(hdr, text="● 초기화 중...",
                             font=ctk.CTkFont("Courier New",12),
                             text_color="#FFB347")
status_label.pack(side="right", padx=20)

main = ctk.CTkFrame(root, fg_color="transparent")
main.pack(fill="both", expand=True, padx=8, pady=8)

# ════════════════════════════════
# 왼쪽: 버튼 + 로그
# ════════════════════════════════
LP = ctk.CTkFrame(main, fg_color="#0B1228", corner_radius=12, width=300)
LP.pack(side="left", fill="y", padx=(0,5))
LP.pack_propagate(False)

def _sepL(): ctk.CTkFrame(LP, fg_color="#1A2A50", height=1).pack(fill="x", padx=10, pady=4)
def _secL(t): ctk.CTkLabel(LP, text=t, font=ctk.CTkFont("Courier New",10,"bold"),
                            text_color="#00CCEE").pack(padx=12, pady=(4,2), anchor="w")

_secL("▶  로봇 제어");  _sepL()

BTN = dict(font=ctk.CTkFont("Courier New",11,"bold"), corner_radius=8, height=40, border_width=1)
for lbl, cmd, fg, bg in [
    ("🏠  홈으로 이동",            lambda: run_task(move_home),              "#00D4FF","#001A30"),
    ("⬜  영점으로 이동",           lambda: run_task(move_zero),              "#44FF88","#001A15"),
    ("🔄  작업1: 2×2 Pick&Place",  lambda: run_task(task_pick2by2_place2by2),"#FF8C42","#1A0E00"),
    ("📦  작업2: 팔레타이징",        lambda: run_task(task_pal_2by2_2layer),   "#FFD700","#1A1500"),
    ("⛔  즉시 정지",               stop_immediate,                           "#FF3366","#200010"),
    ("🔌  연결 해제 및 종료",        shutdown,                                 "#778899","#101418"),
]:
    ctk.CTkButton(LP, text=lbl, command=cmd, fg_color=bg, hover_color=fg,
                  text_color=fg, border_color=fg, **BTN).pack(padx=12, pady=2, fill="x")

_sepL()
manual_btn = ctk.CTkButton(LP, text="🎚  수동 OFF  (실시간 폴링)",
                            command=toggle_manual_mode,
                            fg_color="#1A2A50", hover_color="#2A3A60",
                            text_color="#88BBFF", border_color="#334466", **BTN)
manual_btn.pack(padx=12, pady=2, fill="x")

_sepL();  _secL("▶  실시간 로그")
log_area = ctk.CTkTextbox(LP, font=ctk.CTkFont("Courier New",9),
                           text_color="#77AACC", fg_color="#04070D",
                           border_color="#1A2A50", border_width=1, state="disabled")
log_area.pack(padx=10, pady=(0,10), fill="both", expand=True)
sys.stdout = TextHandler(log_area)

# ════════════════════════════════
# 중앙: 3D 뷰
# ════════════════════════════════
CP = ctk.CTkFrame(main, fg_color="#04070D", corner_radius=12)
CP.pack(side="left", fill="both", expand=True, padx=5)

fig_3d = Figure(figsize=(6,6), dpi=90)   # dpi 낮춰서 가볍게
fig_3d.patch.set_facecolor('#06090F')
ax_3d  = fig_3d.add_subplot(111, projection='3d')

canvas_3d = FigureCanvasTkAgg(fig_3d, master=CP)
cw = canvas_3d.get_tk_widget()
cw.configure(bg='#06090F', highlightthickness=0)
cw.pack(fill="both", expand=True, padx=6, pady=6)

ib = ctk.CTkFrame(CP, fg_color="#0B1228", corner_radius=8, height=32)
ib.pack(fill="x", padx=6, pady=(0,6))
ib.pack_propagate(False)
ee_info_label = ctk.CTkLabel(ib,
    text="EE  X=+0.0000 m    Y=+0.0000 m    Z=+0.0000 m",
    font=ctk.CTkFont("Courier New",11), text_color="#00CCEE")
ee_info_label.pack(side="left", padx=12)
mode_indicator = ctk.CTkLabel(ib, text="● POLLING",
    font=ctk.CTkFont("Courier New",10,"bold"), text_color="#44FF88")
mode_indicator.pack(side="right", padx=12)

# ════════════════════════════════
# 오른쪽: 관절 슬라이더
# ════════════════════════════════
RP = ctk.CTkFrame(main, fg_color="#0B1228", corner_radius=12, width=280)
RP.pack(side="left", fill="y")
RP.pack_propagate(False)

def _sepR(): ctk.CTkFrame(RP, fg_color="#1A2A50", height=1).pack(fill="x", padx=10, pady=4)
def _secR(t): ctk.CTkLabel(RP, text=t, font=ctk.CTkFont("Courier New",10,"bold"),
                            text_color="#00CCEE").pack(padx=12, pady=(6,2), anchor="w")

_secR("▶  관절각 / 수동 제어");  _sepR()

SCOLS = ['#00D4FF','#44FF88','#FF8C42','#FFD700','#FF3366','#CC88FF']

for i in range(6):
    hrow = ctk.CTkFrame(RP, fg_color="transparent")
    hrow.pack(fill="x", padx=12, pady=(7,1))
    ctk.CTkLabel(hrow, text=f"J{i+1}",
                 font=ctk.CTkFont("Courier New",12,"bold"),
                 text_color=SCOLS[i], width=30).pack(side="left")
    lbl = ctk.CTkLabel(hrow, text=f"{HOME_ANGLES_DEG[i]:+7.2f}°",
                       font=ctk.CTkFont("Courier New",12,"bold"),
                       text_color="#DDEEFF")
    lbl.pack(side="right")
    angle_labels.append(lbl)

    sl = ctk.CTkSlider(RP, from_=-180, to=180, height=18,
                       command=lambda v, idx=i: _on_slider(v, idx),
                       button_color=SCOLS[i], progress_color=SCOLS[i],
                       fg_color="#1A2A50")
    sl.set(HOME_ANGLES_DEG[i])
    sl.pack(fill="x", padx=12, pady=(1,0))
    joint_sliders.append(sl)

    gb = ctk.CTkProgressBar(RP, height=5, corner_radius=2,
                             progress_color=SCOLS[i], fg_color="#111A2A")
    gb.set((HOME_ANGLES_DEG[i]+180)/360)
    gb.pack(fill="x", padx=12, pady=(2,0))
    gauge_bars.append(gb)

_sepR()
ctk.CTkLabel(RP, text="  Range: ±180°  |  Red=X  Green=Y  Blue=Z",
             font=ctk.CTkFont("Courier New",8), text_color="#334455"
             ).pack(anchor="w", padx=12)
_sepR()
ctk.CTkButton(RP, text="↺  홈 포즈 리셋",
              command=lambda: _apply_angles_to_ui(HOME_ANGLES_DEG),
              fg_color="#0D1E38", hover_color="#1A3060",
              text_color="#88AACC", border_color="#224466",
              font=ctk.CTkFont("Courier New",10,"bold"),
              corner_radius=8, height=36, border_width=1
              ).pack(padx=12, pady=4, fill="x")

# ═══════════════════════════════════════════
# 초기화
# ═══════════════════════════════════════════
draw_robot_3d(HOME_ANGLES_DEG)
root.after(120, process_robot_queue)

if not SIMULATION_MODE and indy:
    try:
        indy.connect()
        print("로봇 연결 성공")
        set_status("연결됨 – 실시간 폴링 중", "#44FF88")
        threading.Thread(target=polling_thread, daemon=True).start()
    except Exception as e:
        print(f"연결 실패: {e}")
        set_status("연결 실패 – 시뮬레이션", "#FF4444")
else:
    print("[시뮬레이션 모드] 슬라이더로 자세 확인 가능")
    set_status("시뮬레이션 모드", "#FFB347")

root.protocol("WM_DELETE_WINDOW", shutdown)
root.mainloop()
```
